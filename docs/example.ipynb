# Example usage

Here we are showing an example of how to use `pylyrics` to extract and analyze lyrics. With pylrics you can download songs attribute datasets from Kaggle, extract lyrics and generate a word cloud.

# Imports

```python
from pylyrics import extract_lyrics as el
from pylyrics import plot_cloud as pc
from pylyrics import download_data as dd
from pylyrics import clean_text as ct
```

# Downloading the data

```python
dd.download_data("geomack/spotifyclassification", "data/spotify_attributes", ["song_title", "artist"])
```

# Extracting a lyrics 

```python
lyrics = el.extract_lyrics("22", "Taylor Swift")
lyrics
```

# Cleaning the text

```python
ct.clean_text(lyrics)
```

# Making the wordcloud

```python
song = { "Taylor Swift": "22"}
file_path = "data/wordcloud"
pc.plot_cloud(song, file_path, max_font_size=30, max_words=100, background_color='black')
```
